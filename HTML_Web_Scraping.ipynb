Findall with Pandas

```python
import pandas as pd
```

```python
url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets'

r = requests.get(url)

soup = BeautifulSoup(r.text, 'lxml')

# Get the name and create a list
names = soup.find_all('a',{'class':'title'})
#print(names)
product_name = []

for i in names:
  name = i.text
  product_name.append(name)

print(product_name)

# Get the prices and create a list
prices = soup.find_all('h4',{'class':'pull-right price'})
product_price = []
for i in prices:
  price = i.text
  product_price.append(price)

print(product_price)

# Get the descriptions and create a list
descs = soup.find_all('p',{'class':'description'})
product_desc = []
for i in descs:
  desc = i.text
  product_desc.append(desc)

print(product_desc)

# Get the reviews and create a list
reviews = soup.find_all('p',{'class':'pull-right'})
product_review = []
for i in reviews:
  review = i.text
  product_review.append(review)

print(product_review)

# Create a dataframe with all the previus data
df = pd.DataFrame({'Product Name':product_name,
                   'Prices':product_price,
                   'Description':product_desc,
                   'Reviews':product_review})
df
```

Extract Data From Tables

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.oneindia.com/coronavirus-affected-cities-districts-in-india.html'
r = requests.get(url)
soup = BeautifulSoup(r.text, 'lxml')

# Find the table
table = soup.find('table', {'class': 'corona-cases-table'})

# Get the table headers
titles = []
headers = table.find_all('th')
for header in headers:
    title = header.text
    titles.append(title)

# Create the dataframe
df = pd.DataFrame(columns=titles)

# Get the table content and append it to the dataframe
rows = table.find_all('tr')

for row in rows[1:]:
    data = row.find_all('td')
    row_values = [tr.text for tr in data]

    # Check for mismatched columns
    if len(row_values) == len(df.columns):
        df.loc[len(df)] = row_values
    else:
        print(f"Skipping row: {row_values}")

df
```

```python
url = 'https://ticker.finology.in/'

r = requests.get(url)

#print(r)

soup = BeautifulSoup(r.text,'lxml')

table = soup.find('table',{'class':'table table-sm table-hover screenertable'})
# Get tables headers
titles = []
headers = table.find_all('th')
for i in headers:
  title = i.text
  titles.append(title)

#print(titles)
df = pd.DataFrame(columns=titles)
#print(df)

# Let´s obtain all tr tags to get the content of the table
rows = table.find_all('tr')

for i in rows[1:]: # it begans with 1 because we already have the 0 index (it is the headers)
  data = i.find_all('td')
  row = [tr.text for tr in data]
  #print(row)
  l = len(df)
  df.loc[l] = row

print(df)
```

```python
for i in range(1,11):
  url = 'https://www.flipkart.com/search?q=mobile+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='+str(i)
  r = requests.get(url)
  #print(r)

  soup = BeautifulSoup(r.text,'lxml')
  box = soup.find('div',{'class':'_1YokD2 _3Mn1Gg'}) # use box because we eant to get the data from the box and not from the bottom page Reviews for Popular Mobiles
  #print(soup)

  # Get the product name and create a list
  names = box.find_all('div',{'class':'_4rR01T'})
  #print(names)

  product_name = []

  for i in names:
    name = i.text
    product_name.append(name)

  #print(len(product_name))

  # Get the price of the product
  prices = box.find_all('div',{'class':'_30jeq3 _1_WHN1'})
  #print(prices)

  price_product = []

  for i in prices:
    price = i.text
    price_product.append(price)

  #print(len(price_product))

  # Get the Description
  descriptions = box.find_all('ul',{'class':'_1xgFaf'})
  #print(descriptions)

  description_product = []

  for i in descriptions:
    description = i.text
    description_product.append(description)

  #print(len(description_product))

  # Get the Reviews
  reviews = box.find_all('div',{'class':'_3LWZlK'})
  review_product = [review.text if review is not None else None for review in reviews]

  #print(len(review_product))

# Create DataFrame
df = pd.DataFrame({
    'Product': product_name,
    'Price': price_product,
    'Description': description_product,
    'Review': review_product
})

df
```
